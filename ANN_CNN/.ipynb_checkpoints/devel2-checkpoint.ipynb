```python
import os
import json
import time
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from core.data_processor import DataLoader

configs = json.load(open('config.json', 'r'))

data = DataLoader(
    os.path.join('../data', configs['data']['filename']),
    configs['data']['train_test_split'],
    configs['data']['columns']
)
```

```python
dataframe = pd.read_csv(os.path.join('../data', configs['data']['filename']), delim_whitespace=True)
```

```python
print(dataframe)
```

```python
x, y = data.get_train_data(
    seq_len=configs['data']['sequence_length'],
    normalise=configs['data']['normalise']
)
```

```python
for i in range(len(x)):
	print(x[i], y[i])
```

```python
x_test, y_test, p0_vec = data.get_test_data(
    seq_len=configs['data']['sequence_length'],
    normalise=configs['data']['normalise']
)
```

```python
for i in range(len(x_test)):
	print(x_test[i], y_test[i])
```

```python
# -- Model Fitting: -- 
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
```

```python
# Get the number of features from the dataset:
n_features = x.shape[2]
print(n_features)
```

```python
n_steps = configs['data']['sequence_length'] - 1
print(n_steps)
```

```python
# Define model
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(n_steps, n_features)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
# Fit model
model.fit(x, y, epochs=100, verbose=0)
```

```python
# Get predictions: 
yhat = model.predict(x_test, verbose=0)
```

```python
# Visualise results for normalised data:
plt.plot(y_test, label='True Data')
plt.plot(yhat, label='Prediction')
plt.legend()
plt.show()
```

```python
# Denormalise:
def denorm_transform(p0_vec, n_pred, n_true):
    p_pred = np.multiply(p0_vec, (n_pred+1))
    p_true = np.multiply(p0_vec, (n_true+1))
    return p_pred,p_true
```

```python
p_pred, p_true = denorm_transform(p0_vec, yhat, y_test)
```

```python
# Visualise results for de-normalised data:
plt.plot(p_true, label='True Data')
plt.plot(p_pred, label='Prediction')
plt.legend()
plt.show()
```

```python
from core.eval_metrics import EvalMetrics
```

```python
# Compute evaluation metrics
assess = EvalMetrics(p_true, p_pred)
MAE    = assess.get_MAE()
RMSE   = assess.get_RMSE()
print("MAE on validation set is: %f" % MAE)
print("RMSE on validation set is: %f" % RMSE)
```
